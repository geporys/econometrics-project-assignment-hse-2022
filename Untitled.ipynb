```python
import pandas as pd
```

```python
df = pd.read_csv('transactions-2023-01-11.csv')
```

```python
# Drop high cardinality columns
df = df.drop(columns=["Transaction Number", "Property ID", "Transaction Size (sq.m)", "Parking", "Project"])
# Drop low-cardinality columns
df = df.drop(columns=["Registration type", "Is Free Hold?", "Master Project"])
# Drop leaky columns
df = df.drop(columns=["Transaction sub type", "Property Type", "Room(s)", "No. of Buyer", "No. of Seller"])
```

```python
df["Transaction Date"] = pd.to_datetime(df["Transaction Date"])
df
```
